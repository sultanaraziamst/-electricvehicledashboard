```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("IEA Global EV Data 2024 new.csv")
df.head(10)
```

```python
df.info
```

```python
df['value'] = pd.to_numeric(df['value'], errors='coerce') 
df['percentage'] = pd.to_numeric(df['percentage'], errors='coerce') 
```

```python
year_count = df['year'].count()
year_mean = df ['year'].mean()
year_median = df['year'].median()
year_mode = df ['year'].mode()[0]
year_range = df['year'].max() -df['year'].min()

print(f"Count: {year_count}, Mean:{year_mean}, Median: {year_median}, Mode: {year_mode}, range: {year_range}")

```

```python
region_frequncy = df['region'].value_counts()
print(region_frequncy)
```

```python
catergory_frequency = df['category'].value_counts()
print(catergory_frequency)
```

```python
mode_poportion = df ['mode'].value_counts(normalize = True)
print(mode_poportion)
```
